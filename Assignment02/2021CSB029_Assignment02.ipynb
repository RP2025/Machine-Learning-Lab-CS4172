***i] Download https://archive.ics.uci.edu/dataset/17/breast+cancer+wisconsin+diagnostic Data Set (already in the needed format).
The dataset is used to recognize 2 types of cancer to be predicted (benign or
malignant).***

```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
breast_cancer_wisconsin_diagnostic = fetch_ucirepo(id=17)

# data (as pandas dataframes)
X = breast_cancer_wisconsin_diagnostic.data.features
y = breast_cancer_wisconsin_diagnostic.data.targets

# metadata
print(breast_cancer_wisconsin_diagnostic.metadata)

# variable information
print(breast_cancer_wisconsin_diagnostic.variables)
```

```python
cols = ["ID","Diagnosis",
        "radius1","texture1","perimeter1","area1","smoothness1","compactness1","concavity1","concave_points1","symmetry1","fractal_dimension1",
        "radius2","texture2","perimeter2","area2","smoothness2","compactness2","concavity2","concave_points2","symmetry2","fractal_dimension2",
        "radius3","texture3","perimeter3","area3","smoothness3","compactness3","concavity3","concave_points3","symmetry3","fractal_dimension3"]
df = pd.read_csv("wdbc.data",names=cols)
```

```python
df.head(10)
```

```python
df.drop(["ID"],axis=1,inplace=True)
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
X = df.drop(["Diagnosis"],axis = 1)
X.head(10)
```

```python
Y = df[["Diagnosis"]]
Y['Diagnosis'] = Y['Diagnosis'].replace({'B': 0, 'M': 1})
Y.head(10)
```

***ii] Implement Logistic Regression using scikit-learn package.***

***a] Use ‘newton-cg’, ‘lbfgs’, ‘liblinear’ solver for the regression model.***

***b] For each solver use ‘l1’, ‘l2’, ‘none’ penalty to train the model.***

***c] Split the dataset in to 80:10:10 percent (train : validation : test) (use seed = 5 for splitting).***

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_temp, Y_train, Y_temp = train_test_split(X, Y, test_size=0.20, random_state=5)
X_val, X_test, Y_val, Y_test = train_test_split(X_temp, Y_temp, test_size=0.50, random_state=5)
```

```python
print(X_train.shape)
print(X_val.shape)
print(X_test.shape)
score = []
```

***d] Train the model initially with 80% training data and create a table for the coefficients of all the features.***

```python
lr1 = LogisticRegression(
    penalty='l2',
    solver='newton-cg'
)
lr1.fit(X_train,Y_train)
score.append(lr1.score(X_train,Y_train))
score[-1]
```

```python
item = lr1.coef_[0].tolist()
item = ['newton-cg','l2'] + item
table = pd.DataFrame(data = np.array([item]),columns=["solver","loss"]+cols[2:])
```

```python
lr2 = LogisticRegression(
    penalty='none',
    solver='newton-cg'
)
lr2.fit(X_train,Y_train)
score.append(lr2.score(X_train,Y_train))
score[-1]
```

```python
item = lr2.coef_[0].tolist()
item = ['newton-cg','none'] + item
table = pd.concat([table,pd.DataFrame(data = np.array([item]),columns=["solver","loss"]+cols[2:])])
```

```python
lr3 = LogisticRegression(
    penalty='l2',
    solver='lbfgs'
)
lr3.fit(X_train,Y_train)
score.append(lr3.score(X_train,Y_train))
score[-1]
```

```python
item = lr3.coef_[0].tolist()
item = ['lbfgs','l2'] + item
table = pd.concat([table,pd.DataFrame(data = np.array([item]),columns=["solver","loss"]+cols[2:])])
```

```python
item = lr4.coef_[0].tolist()
item = ['lbfgs','none'] + item
table = pd.concat([table,pd.DataFrame(data = np.array([item]),columns=["solver","loss"]+cols[2:])])
```

```python
lr5 = LogisticRegression(
    penalty='l2',
    solver='liblinear'
)
lr5.fit(X_train,Y_train)
score.append(lr5.score(X_train,Y_train))
score[-1]
```

```python
item = lr5.coef_[0].tolist()
item = ['liblinear','l2'] + item
table = pd.concat([table,pd.DataFrame(data = np.array([item]),columns=["solver","loss"]+cols[2:])])
```

```python
grid_search = GridSearchCV(
    lr1,
    param_grid,
    scoring='accuracy'
)
grid_search.fit(X_train_final, Y_train_final)
best_model = grid_search.best_estimator_
print("Best hyperparameters found lr1: ", grid_search.best_params_)
best_param[0] = [grid_search.best_params_['C'],grid_search.best_params_['max_iter']]
```
